```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(12, 15))

# Colors for each race
race_colors = {
    "White": "blue",
    "Black": "green",
    "Hispanic": "red",
    "Asian": "purple",
    "Indigenous": "orange",
}

# Drawing strip plots separately for each race to ensure unique color per race and two entries per disease
for race in race_colors:
    # Filter the data for each race
    race_data = df_long[df_long["Race"] == race]
    sns.stripplot(
        data=race_data,
        x="Rank",
        y="Disease Type",
        hue="Type",
        dodge=True,
        jitter=False,
        marker="o",
        palette=[race_colors[race]],
        linewidth=1,
        edgecolor="w",
        size=10,
        orient="h",
    )

# Customizing the plot
ax.set(xlim=(0, 6), xlabel="Rank", ylabel="")
ax.set_title(
    "Comparison of Disease Rankings Between Pile and Real-World Data by Race",
    fontsize=16,
)
ax.legend(title="Data Type")

# Remove existing race legends and add a custom legend for races
handles, labels = ax.get_legend_handles_labels()
legend_labels = list(race_colors.keys())
custom_handles = [
    plt.Line2D(
        [0], [0], color=race_colors[race], marker="o", linestyle="", markersize=10
    )
    for race in race_colors
]
ax.legend(custom_handles, legend_labels, title="Race", bbox_to_anchor=(1.05, 1), loc=2)

# Enhance visibility
sns.despine(left=True, bottom=True)

# Show the plot
plt.show()
```
