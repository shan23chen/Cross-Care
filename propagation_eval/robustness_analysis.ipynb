```python
def most_common_list(series):
    # Convert each list from string representation to actual list
    series = series.apply(lambda x: eval(x))
    # Counting each list (converted to a tuple for hashability)
    counts = Counter(series.apply(tuple))
    # Finding the most common list and its count
    most_common, count = counts.most_common(1)[0]
    # Converting the tuple back to a list for the result
    return count, list(most_common)

# Apply the function to each group of 'disease' and 'model_name'
result = df.groupby(['disease', 'model_name'])['numeric_ranks'].apply(lambda x: most_common_list(x)).to_dict()
```

```python
import pandas as pd
from collections import Counter

# Assuming data_ranking is the DataFrame containing the template_ranking.csv data
unique_models = df['model_name'].unique()

model_analysis = {}

for model in unique_models:
    # Filtering data for the current model
    model_data = {key: value for key, value in result.items() if model in key[1]}
    
    # If there's no data for the model, skip to the next one
    if not model_data:
        continue
    
    # Extracting 'Max Count of Same Numeric Ranks' directly
    max_counts = [value[0] for value in model_data.values()]
    
    # Calculate the range (min and max) and frequencies of "Max Count of Same Numeric Ranks"
    min_max_count_model = min(max_counts), max(max_counts)
    frequencies_model = Counter(max_counts)
    
    # Storing the results
    model_analysis[model] = {
        'Range': min_max_count_model,
        'Frequencies': frequencies_model
    }

# To display a sample of the analysis, let's print the analysis for one model
sample_model = list(model_analysis.keys())[0]  # Taking the first model for sample
print(f"Analysis for model: {sample_model}")
print("Range of 'Max Count of Same Numeric Ranks':", model_analysis[sample_model]['Range'])
print("Frequencies of 'Max Count of Same Numeric Ranks':", model_analysis[sample_model]['Frequencies'])
```

```python
# To compare the frequency distributions of the "Max Count of Same Numeric Ranks" among the models,
# let's create a summary DataFrame that compiles this information for easier comparison.

# Initializing a list to hold compiled frequency data for each model
compiled_frequencies = []

# Iterate through each model's analysis result to compile frequency data
for model, analysis in model_analysis.items():
    for rank, freq in analysis['Frequencies'].items():
        compiled_frequencies.append({'Model': model, 'Max Count of Same Numeric Ranks': rank, 'Frequency': freq})

# Convert the compiled list to a DataFrame for visualization
frequency_df = pd.DataFrame(compiled_frequencies)

# Pivot the DataFrame to have one row per model and columns for each "Max Count of Same Numeric Ranks", filled with frequencies
pivot_frequency_df = frequency_df.pivot_table(index='Model', columns='Max Count of Same Numeric Ranks', values='Frequency', fill_value=0)

pivot_frequency_df
```

```python
# split df based on template from 0-9 and 10-19
df_0_9 = df[df['template'] <= 9]
df_10_19 = df[df['template'] >= 10]
```

```python
from collections import defaultdict

def count_top_bottom_matches_corrected(data):
    # Assuming 'ranked_demographics' are already converted to lists
    # Initialize a dictionary to hold the counts for each model-disease
    top_bottom_matches_corrected = {}
    
    # Group by model and disease
    grouped = data.groupby(['model_name', 'disease'])
    
    for (model, disease), group in grouped:
        # Lists to hold top (first) and bottom (last) demographics across templates
        top_demographics = []
        bottom_demographics = []
        
        # Collect the top and bottom demographics for each template
        for _, row in group.iterrows():
            demographics = row['ranked_demographics']
            if demographics:  # Ensure the list is not empty
                top_demographics.append(demographics[0])
                bottom_demographics.append(demographics[-1])
        
        # Count how many times the top and bottom demographics are the same
        top_count = Counter(top_demographics)
        bottom_count = Counter(bottom_demographics)       
        top1 = top_count.most_common(1)[0][0]
        bottom1 = bottom_count.most_common(1)[0][0]
        # Store the result
        top_bottom_matches_corrected[(model, disease)] = [top_count, bottom_count, top1, bottom1]
    
    return top_bottom_matches_corrected

# Make sure 'ranked_demographics' column is correctly formatted as lists
# df['ranked_demographics'] = df['ranked_demographics'].apply(eval)  # Only if not already lists

# Apply the corrected function
top_bottom_matches_corrected = count_top_bottom_matches_corrected(df)

# Output the corrected results
top_bottom_matches_corrected

```

```python
def convert_tb(tb):
    # top_bottom_matches_corrected = pd.DataFrame(tb)
    # Convert the dictionary into a DataFrame
    df_top_bottom = pd.DataFrame.from_dict(tb, orient='index', columns=['top_count', 'bottom_count','top1', 'bottom1'])

    # Reset the index
    df_top_bottom = df_top_bottom.reset_index()

    # Rename the index columns
    return df_top_bottom

df_tb0 = convert_tb(tb0)
df_tb1 = convert_tb(tb1)
```

```python
def convert_tb(df_top_bottom):
    modelname = []
    disease = []
    for i in df_top_bottom['index']:
        temp = list(i)
        temp[0] = temp[0]
        temp[1] = temp[1]
        modelname.append(temp[0])
        disease.append(temp[1])

    df_top_bottom['model_name'] = modelname
    df_top_bottom['disease'] = disease
    return df_top_bottom

df_tb0 = convert_tb(df_tb0)
df_tb1 = convert_tb(df_tb1)
```

```python
# iterating dieases and check how many times top1 and bottom1 are same across models
def top1_bottom1_same(df_top_bottom):
    disease = set(df_top_bottom['disease'])
    disease

    top1_bottom1_same = []
    for i in disease:
        temp = df_top_bottom[df_top_bottom['disease'] == i]
        top1 = temp['top1']
        bottom1 = temp['bottom1']
        top1_bottom1_same.append({i:Counter(top1)})
    return top1_bottom1_same


top1_bottom1_same(df_tb0)
```
